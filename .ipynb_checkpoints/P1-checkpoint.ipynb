## Helper Functions

Below are some helper functions!

```python
import math
preFrame = []

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=15):
    """
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    global preFrame # the line information from previous image
    rlaneX = [] # right lane X
    rlaneY = [] # right lane Y
    llaneX = [] # left lane X
    llaneY = [] # left lane Y
    
    lslopes = []
    rslopes = []
    
    # finds the lines for left and right lines based on their slopes
    # adds their coordiantes the the lists
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
            if slope>0.5 and slope<1:
                rslopes.append(slope)
                rlaneX.extend((x1,x2))
                rlaneY.extend((y1,y2))
            elif slope>-1 and slope<-0.5:
                lslopes.append(slope)
                llaneX.extend((x1,x2))
                llaneY.extend((y1,y2))
    
    # When no Hough lines is found OR if the variance of the slopes is more than usual use the previous lines
    if len(rlaneX)==0 or len(llaneX)==0 or np.var(lslopes)>0.005 or np.var(rslopes)>0.005:
        if len(preFrame)!=0:
            cv2.line(img, (int(preFrame[0]), int(preFrame[1])), 
                          (int(preFrame[2]), int(preFrame[3])), color, thickness)
            cv2.line(img, (int(preFrame[4]), int(preFrame[5])), 
                          (int(preFrame[6]), int(preFrame[7])), color, thickness)
        return img
    
    # Fits a first order polynomial (line) to the coordiantes of the Hough lines in left and right lines list
    lfit = np.polyfit(llaneX,llaneY,1)
    rfit = np.polyfit(rlaneX,rlaneY,1)
    
    imshape = img.shape
    
    # Y coordinates of the left and right lines for drwaing
    lBottomY = imshape[0]
    rBottomY = imshape[0] 
    lTopY = imshape[0]*2/3
    rTopY = imshape[0]*2/3
    
    # Clculating the Xs of the lines for drawing
    # y = mx + b, x = (y-b)/m
    lBottomX = (lBottomY - lfit[1])/lfit[0]
    rBottomX = (rBottomY - rfit[1])/rfit[0]
    lTopX = (lTopY - lfit[1])/lfit[0]
    rTopX = (rTopY - rfit[1])/rfit[0]
    
    ## Checking if the lines intersect at top. if yes finds the intersection
    if (lTopX>rTopX):
        lTopX =  (lfit[1] - rfit[1])/(rfit[0] - lfit[0]) 
        rTopX = lTopX
        lTopY = (rfit[0]*lfit[1] - lfit[0]*rfit[1])/(rfit[0] - lfit[0])
        rTopY = lTopY
    
    # adds the coordinates of the left and right lines to the currentFrame
    currentFrame = np.array([lBottomX, lBottomY, 
                             lTopX, lTopY, 
                             rBottomX, rBottomY, 
                             rTopX, rTopY])
    # To have smoothe detection of the lines in a video, the coordinates of the previous frame is contributed
    if (len(preFrame) != 0):
        currentFrame = 0.7 * currentFrame + 0.3 * preFrame
    preFrame = currentFrame 
    
    # Adding the lines to the frame
    cv2.line(img, (int(currentFrame[0]), int(currentFrame[1])), 
                  (int(currentFrame[2]), int(currentFrame[3])), color, thickness)
    cv2.line(img, (int(currentFrame[4]), int(currentFrame[5])), 
                  (int(currentFrame[6]), int(currentFrame[7])), color, thickness)
    
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, connected= True):
    """        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

## Pipeline:
In the folllowing cell, the pipeline is constructed as a function with the following steps:

* Conversion to grayscale
* Masking with white and yellow color mask
* Gaussian blur with kernel 5
* Image is fed into Canny edge detection function
* the edges are filtered in a region of interest
* 

```python
## Setting up the constants
yellow_low = np.array([20, 100, 100], dtype = "uint8") # Low yellow color for masking
yellow_high = np.array([30, 255, 255], dtype="uint8") # High yellow color for masking 
low_threshold = 50 # canny low threshhol
high_threshold = 150 # canny high threshhold
rho = 2 # distance resolution in pixels of the Hough grid
theta = np.pi/180 # angular resolution in radians of the Hough grid
threshold = 15     # minimum number of votes (intersections in Hough grid cell)
min_line_len = 40 #minimum number of pixels making up a line
max_line_gap = 20    # maximum gap in pixels between connectable line segments

## image processing function
def process_image(image):
    gray = grayscale(image)
    
    # Masking yellow and white pixels
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    yellow_mask = cv2.inRange(hsv, yellow_low, yellow_high)
    white_mask = cv2.inRange(gray, 150, 255)
    color_mask = cv2.bitwise_or(white_mask, yellow_mask)
    color_masked_gray = cv2.bitwise_and(gray, color_mask)
    #
    # Bluring the image
    blur_gray = gaussian_blur(color_masked_gray, 5)
    # 
    edges = canny(blur_gray, low_threshold, high_threshold)
    
    imshape = image.shape
    vertices = np.array([[(0,imshape[0]),
                          (imshape[1]*19/40, imshape[0]*6/10),
                          (imshape[1]*21/40, imshape[0]*6/10), 
                          (imshape[1],imshape[0])]], 
                          dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)
    lines = hough_lines(masked_edges, rho, theta, threshold, min_line_len, max_line_gap);
    result = weighted_img(lines, image, α=0.8, β=1., λ=0.)
    return result
```

## Test Images

The test images are loaded and the pipline is appled to them. The outputs are:

```python
images = os.listdir("test_images/")
imageDir = "test_images/"
outDir = "test_images_output/"

for image in images: 
    preFrame = []
    img = mpimg.imread(imageDir+image)
    pro_img = process_image(img)
    mpimg.imsave(outDir+image, pro_img)
    plt.figure()
    plt.imshow(pro_img)
```
